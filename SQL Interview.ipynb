```python
import pandas as pd
import sqlite3

#pd.read_csv?

pd.read_csv("url.zip", nrows=10, chunksize=100_000)

db = sqlite3.connect('citibike.db')

pd.DataFrame.to_sql
```

```python
url= ""

for i in range(5,6,7,8,9):
    print(url % i)
    chunks=pd.read_csv(url,nrows=10, chunksize=100_00)
    for chunk in chunks:
        chunk.columns = [column.replace(' ','_') for column in chunk.columns]
        chunk.to_sql('tripdata',db, if_exists='append')        
```

```python
res = Q("""
select (2017-birth_year) as age , count(*) from tripdata group by age having age>1
""")
res
```

```python
res = Q("""
select bikeid, sum(tripduration)/3600 from tripdata group by bikeid order by 2 
""")
res # order by 2 means position of column. 3600 means it will return hours
```
